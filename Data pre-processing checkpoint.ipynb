```python
# Select your strategy to handle missing values, and tell us why you had made that choice.

# The distribution of the counter number is not normally distributed has the mean is far greater than median
df['counter_number'] = df['counter_number'].fillna(value=df['counter_number'].median())

# The distribution of the reading_remarque is slightly distributed has the mean and the median are slightly close 
df['reading_remarque'] = df['reading_remarque'].fillna(value=df['reading_remarque'].mean())
```
